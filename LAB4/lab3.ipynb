## Lý thuyết
Thuật toán di truyền (Genetic Algorithm - GA) là một phương pháp tìm kiếm dựa trên cơ chế tiến hóa tự nhiên, lấy cảm hứng từ thuyết tiến hóa của Darwin. GA được sử dụng để tìm lời giải gần tối ưu cho các bài toán tối ưu hóa phức tạp.

### Các bước chính của thuật toán:
1. **Khởi tạo quần thể**: Tạo ngẫu nhiên một tập hợp các cá thể (giải pháp tiềm năng).
2. **Đánh giá độ thích nghi (Fitness Evaluation)**: Tính giá trị hàm mục tiêu cho mỗi cá thể.
3. **Lựa chọn (Selection)**: Chọn các cá thể tốt nhất để lai ghép (thường sử dụng các phương pháp như Tournament Selection hoặc Roulette Wheel).
4. **Lai ghép (Crossover)**: Kết hợp hai cá thể để tạo ra cá thể mới.
5. **Đột biến (Mutation)**: Thay đổi ngẫu nhiên một số gen để duy trì tính đa dạng.
6. **Lặp lại**: Quay lại bước 2 cho đến khi đạt điều kiện dừng (số thế hệ hoặc giá trị fitness mong muốn).

Ngoài ra, chúng ta sẽ trực quan hóa sự thay đổi của giá trị fitness qua các thế hệ.


## Ví dụ 1: Tối ưu hóa hàm một biến

### Bài toán  
Tìm $x$ sao cho  
$$
f(x) = -(x^2) + 10x + 50
$$  
đạt giá trị lớn nhất trong khoảng  
$$
x \in [0, 10].
$$

# Mã giả
```plaintext
Tạo quần thể mới rỗng
Lặp cho đến khi quần thể mới đủ N cá thể:
    Chọn cha mẹ 1 bằng Tournament Selection:
        - Chọn ngẫu nhiên tournament_size cá thể trong quần thể
        - Chọn cá thể có fitness cao nhất làm cha mẹ 1
    Chọn cha mẹ 2 bằng Tournament Selection tương tự
    
    Thực hiện lai ghép (crossover) giữa cha mẹ 1 và cha mẹ 2 với xác suất crossover_rate:
        Nếu crossover thành công:
            Tạo con mới có giá trị trung bình của cha mẹ
        Ngược lại:
            Con mới là cha mẹ 1
            
    Thực hiện đột biến cho con mới với xác suất mutation_rate:
        Nếu đột biến:
            Gán con mới giá trị ngẫu nhiên trong [min_val, max_val]
            
    Thêm con mới vào quần thể mới
    
Cập nhật quần thể hiện tại = quần thể mới
In ra cá thể tốt nhất và giá trị fitness tương ứng của thế hệ


````python
import numpy as np
import random
import matplotlib.pyplot as plt

# Hàm mục tiêu
def fitness_function(x):
    return -(x**2) + 10*x + 50

# Khởi tạo quần thể
def initialize_population(pop_size, min_val, max_val):
    return [random.uniform(min_val, max_val) for _ in range(pop_size)]

# Lựa chọn (Tournament Selection)
def select_parents(population, fitness_values, tournament_size=3):
    selected = random.sample(range(len(population)), tournament_size)
    best_idx = max(selected, key=lambda i: fitness_values[i])
    return population[best_idx]

# Lai ghép
def crossover(parent1, parent2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        return (parent1 + parent2) / 2
    return parent1

# Đột biến
def mutate(individual, mutation_rate=0.1, min_val=0, max_val=10):
    if random.random() < mutation_rate:
        return random.uniform(min_val, max_val)
    return individual

# Thuật toán di truyền
def genetic_algorithm_example1(pop_size=50, generations=100, min_val=0, max_val=10):
    population = initialize_population(pop_size, min_val, max_val)
    best_fitness_history = []
    
    for generation in range(generations):
        fitness_values = [fitness_function(x) for x in population]
        best_fitness = max(fitness_values)
        best_fitness_history.append(best_fitness)
        
        new_population = []
        for _ in range(pop_size):
            parent1 = select_parents(population, fitness_values)
            parent2 = select_parents(population, fitness_values)
            child = crossover(parent1, parent2)
            child = mutate(child)
            new_population.append(child)
        
        population = new_population
        best_individual = population[np.argmax([fitness_function(x) for x in population])]
        print(f'Thế hệ {generation + 1}: x = {best_individual:.4f}, f(x) = {best_fitness:.4f}')
    
    best_fitness = max([fitness_function(x) for x in population])
    best_individual = population[np.argmax([fitness_function(x) for x in population])]
    
    # Trực quan hóa
    plt.plot(range(generations), best_fitness_history)
    plt.xlabel('Thế hệ')
    plt.ylabel('Giá trị f(x) tốt nhất')
    plt.title('Sự thay đổi của f(x) qua các thế hệ')
    plt.show()
    
    return best_individual, best_fitness

# Chạy thuật toán
best_x, best_f = genetic_algorithm_example1()
print(f'\nKết quả cuối cùng: x = {best_x:.4f}, f(x) = {best_f:.4f}')
````

### Mã giả 

```plaintext
Tạo quần thể mới rỗng
Lặp cho đến khi quần thể mới đủ N cá thể:
    Chọn cha mẹ 1 bằng Tournament Selection trên quần thể hiện tại
    Chọn cha mẹ 2 bằng Tournament Selection trên quần thể hiện tại
    Thực hiện lai ghép (crossover) giữa cha mẹ 1 và cha mẹ 2 với xác suất crossover_rate:
        Nếu crossover thành công:
            Tạo con mới bằng trung bình giá trị (x, y) của cha mẹ
        Ngược lại:
            Con mới là cha mẹ 1
    Thực hiện đột biến cho con mới với xác suất mutation_rate từng biến:
        Nếu đột biến:
            Gán giá trị ngẫu nhiên trong [min_val, max_val] cho biến tương ứng
    Thêm con mới vào quần thể mới

Cập nhật quần thể hiện tại = quần thể mới
In kết quả tốt nhất thế hệ

````python
# Hàm mục tiêu
def fitness_function_2d(x, y):
    return -(x**2 + y**2)  # Sử dụng dấu âm vì GA tìm max

# Khởi tạo quần thể
def initialize_population_2d(pop_size, min_val, max_val):
    return [(random.uniform(min_val, max_val), random.uniform(min_val, max_val)) for _ in range(pop_size)]

# Lựa chọn
def select_parents_2d(population, fitness_values, tournament_size=3):
    selected = random.sample(range(len(population)), tournament_size)
    best_idx = max(selected, key=lambda i: fitness_values[i])
    return population[best_idx]

# Lai ghép
def crossover_2d(parent1, parent2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        return ((parent1[0] + parent2[0]) / 2, (parent1[1] + parent2[1]) / 2)
    return parent1

# Đột biến
def mutate_2d(individual, mutation_rate=0.1, min_val=-5, max_val=5):
    x, y = individual
    if random.random() < mutation_rate:
        x = random.uniform(min_val, max_val)
    if random.random() < mutation_rate:
        y = random.uniform(min_val, max_val)
    return (x, y)

# Thuật toán di truyền
def genetic_algorithm_example2(pop_size=50, generations=100, min_val=-5, max_val=5):
    population = initialize_population_2d(pop_size, min_val, max_val)
    best_fitness_history = []
    
    for generation in range(generations):
        fitness_values = [fitness_function_2d(x, y) for x, y in population]
        best_fitness = max(fitness_values)
        best_fitness_history.append(-best_fitness)  # Chuyển về giá trị thực
        
        new_population = []
        for _ in range(pop_size):
            parent1 = select_parents_2d(population, fitness_values)
            parent2 = select_parents_2d(population, fitness_values)
            child = crossover_2d(parent1, parent2)
            child = mutate_2d(child)
            new_population.append(child)
        
        population = new_population
        best_individual = population[np.argmax([fitness_function_2d(x, y) for x, y in population])]
        print(f'Thế hệ {generation + 1}: (x, y) = ({best_individual[0]:.4f}, {best_individual[1]:.4f}), g(x, y) = {-best_fitness:.4f}')
    
    best_fitness = max([fitness_function_2d(x, y) for x, y in population])
    best_individual = population[np.argmax([fitness_function_2d(x, y) for x, y in population])]
    
    # Trực quan hóa
    plt.plot(range(generations), best_fitness_history)
    plt.xlabel('Thế hệ')
    plt.ylabel('Giá trị g(x, y) tốt nhất')
    plt.title('Sự thay đổi của g(x, y) qua các thế hệ')
    plt.show()
    
    return best_individual, -best_fitness

# Chạy thuật toán
best_xy, best_g = genetic_algorithm_example2()
print(f'\nKết quả cuối cùng: (x, y) = ({best_xy[0]:.4f}, {best_xy[1]:.4f}), g(x, y) = {best_g:.4f}')
````

# Bài tập 1
# Thay đổi hàm mục tiêu**: Thử áp dụng thuật toán cho hàm $h(x) = sin(x) + cos(x)$ trong khoảng $x \in [0, 2\pi]$. So sánh kết quả với nghiệm chính xác.

## Tối ưu hàm $h(x) = \sin(x) + \cos(x)$ bằng Thuật Toán Di Truyền (Genetic Algorithm)

## Mục tiêu
- Tối ưu hàm:
  $$
  h(x) = \sin(x) + \cos(x)
  $$
  trong miền:
  $$
  x \in [0, 2\pi]
  $$
- Tìm giá trị **x** sao cho **h(x)** đạt giá trị lớn nhất.

---

## Phân tích nghiệm chính xác

Ta có:
$$
h(x) = \sin(x) + \cos(x) = \sqrt{2} \cdot \sin\left(x + \frac{\pi}{4}\right)
\Rightarrow \max h(x) = \sqrt{2} \approx 1.4142
$$

Xảy ra khi:
$$
x = \frac{\pi}{4} \approx 0.7854
$$

---

## Ý tưởng thuật toán di truyền (GA)

1. **Mã hóa**: Biểu diễn giá trị \( x \) bằng chuỗi nhị phân dài 16 bit.  
2. **Giải mã**: Chuyển từ chuỗi nhị phân sang giá trị thực trong khoảng \([0, 2\pi]\).  
3. **Fitness**: Sử dụng chính giá trị \( h(x) \) làm độ thích nghi.  
4. **Chọn lọc**: Roulette Wheel Selection.  
5. **Lai ghép**: Lai 1 điểm với xác suất cao.  
6. **Đột biến**: Lật bit với xác suất thấp.

---


````python
# code vào  đây nhe
import numpy as np
import matplotlib.pyplot as plt
import random

# Hàm mục tiêu
def fitness_function_1d(x):
    return np.sin(x) + np.cos(x)

# Khởi tạo quần thể
def initialize_population_1d(pop_size, min_val, max_val):
    return [random.uniform(min_val, max_val) for _ in range(pop_size)]

# Lựa chọn
def select_parents_1d(population, fitness_values, tournament_size=3):
    selected = random.sample(range(len(population)), tournament_size)
    best_idx = max(selected, key=lambda i: fitness_values[i])
    return population[best_idx]

# Lai ghép
def crossover_1d(parent1, parent2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        return (parent1 + parent2) / 2
    return parent1

# Đột biến
def mutate_1d(individual, mutation_rate=0.1, min_val=0, max_val=2*np.pi):
    if random.random() < mutation_rate:
        individual = random.uniform(min_val, max_val)
    return individual

# Thuật toán di truyền
def genetic_algorithm_1d(pop_size=50, generations=100, min_val=0, max_val=2*np.pi):
    population = initialize_population_1d(pop_size, min_val, max_val)
    best_fitness_history = []
    
    for generation in range(generations):
        fitness_values = [fitness_function_1d(x) for x in population]
        best_fitness = max(fitness_values)
        best_fitness_history.append(best_fitness)
        
        new_population = []
        for _ in range(pop_size):
            parent1 = select_parents_1d(population, fitness_values)
            parent2 = select_parents_1d(population, fitness_values)
            child = crossover_1d(parent1, parent2)
            child = mutate_1d(child)
            new_population.append(child)
        
        population = new_population
        best_individual = population[np.argmax([fitness_function_1d(x) for x in population])]
        print(f'Thế hệ {generation + 1}: x = {best_individual:.4f}, h(x) = {fitness_function_1d(best_individual):.4f}')
    
    best_fitness = max([fitness_function_1d(x) for x in population])
    best_individual = population[np.argmax([fitness_function_1d(x) for x in population])]
    
    # Trực quan hóa
    plt.plot(range(generations), best_fitness_history)
    plt.xlabel('Thế hệ')
    plt.ylabel('Giá trị h(x) tốt nhất')
    plt.title('Sự thay đổi của h(x) = sin(x) + cos(x)')
    plt.grid(True)
    plt.show()
    
    return best_individual, best_fitness

# Chạy thuật toán
best_x, best_h = genetic_algorithm_1d()
print(f'\nKết quả cuối cùng: x = {best_x:.4f}, h(x) = {best_h:.4f}')

# So sánh nghiệm chính xác
exact_x = np.pi / 4
exact_h = np.sqrt(2)
print(f'Nghiệm chính xác: x = π/4 ≈ {exact_x:.4f}, h(x) = √2 ≈ {exact_h:.4f}')
````

# Bài tập 2: Tối ưu hàm $h(x) = \sin(x) + \cos(x)$ bằng Thuật Toán Di Truyền (Genetic Algorithm)

---

## Mục tiêu

Tối ưu hàm:
$$
h(x) = \sin(x) + \cos(x)
$$
trong khoảng:
$$
x \in [0, 2\pi]
$$

> Tìm giá trị $x$ sao cho $h(x)$ đạt **giá trị lớn nhất**.

---

## Phân tích nghiệm chính xác

Ta có:
$$
h(x) = \sin(x) + \cos(x) = \sqrt{2} \cdot \sin\left(x + \frac{\pi}{4}\right)
\Rightarrow \max h(x) = \sqrt{2} \approx 1.4142
$$

Xảy ra khi:
$$
x = \frac{\pi}{4} \approx 0.7854
$$

---

##  Ý tưởng thuật toán di truyền (GA)

| Thành phần        | Chi tiết                                                                 |
|------------------|-------------------------------------------------------------------------|
| **Mã hóa**        | Dùng chuỗi nhị phân 16 bit để biểu diễn giá trị $x \in [0, 2\pi]$         |
| **Giải mã**       | Ánh xạ nhị phân sang giá trị thực: $x = 0 + \dfrac{\text{decimal}}{2^{16}-1} \cdot 2\pi$ |
| **Fitness**       | Chính là $h(x) = \sin(x) + \cos(x)$                                     |
| **Chọn lọc**      | Roulette Wheel Selection                                                |
| **Lai ghép**      | 1 điểm cắt với xác suất 0.8                                              |
| **Đột biến**      | Xác suất 5% (0.05)                                                       |

---

````python
# code vào đây nhe
# code vào  đây nhe
import numpy as np
import math
import random
import matplotlib.pyplot as plt

# Hàm mục tiêu
def fitness_function(x):
    return math.sin(x) + math.cos(x)  # các phương trình

# Khởi tạo quần thể
def initialize_population(pop_size, min_val, max_val):
    return [random.uniform(min_val, max_val) for _ in range(pop_size)]


# Lựa chọn (Roulette Wheel Selection)
def select_parents(population, fitness_values):
    # Tính tổng fitness
    total_fitness = sum(fitness_values)
    if total_fitness == 0:
        return random.choice(population)  # tránh chia cho 0

    # Xác suất chọn mỗi cá thể
    probabilities = [f / total_fitness for f in fitness_values]
    
    # Chọn một cá thể theo phân phối xác suất
    return random.choices(population, weights=probabilities, k=3)[0]

# Lai ghép
def crossover(parent1, parent2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        return (parent1 + parent2) / 2
    return parent1

# Đột biến
def mutate(individual, mutation_rate=0.05, min_val=0, max_val=2*math.pi):
    if random.random() < mutation_rate:
        return random.uniform(min_val, max_val)
    return individual

# Thuật toán di truyền
def genetic_algorithm_example1(pop_size=50, generations=100, min_val=0, max_val=2*math.pi):
    population = initialize_population(pop_size, min_val, max_val)
    best_fitness_history = []
    
    for generation in range(generations):
        fitness_values = [fitness_function(x) for x in population]
        best_fitness = max(fitness_values)
        best_fitness_history.append(best_fitness)
        
        new_population = []
        for _ in range(pop_size):
            parent1 = select_parents(population, fitness_values)
            parent2 = select_parents(population, fitness_values)
            child = crossover(parent1, parent2)
            child = mutate(child)
            new_population.append(child)
        
        population = new_population
        best_individual = population[np.argmax([fitness_function(x) for x in population])]
        print(f'Thế hệ {generation + 1}: x = {best_individual:.4f}, f(x) = {best_fitness:.4f}')
    
    best_fitness = max([fitness_function(x) for x in population])
    best_individual = population[np.argmax([fitness_function(x) for x in population])]
    
    # Trực quan hóa
    plt.plot(range(generations), best_fitness_history)
    plt.xlabel('Thế hệ')
    plt.ylabel('Giá trị f(x) tốt nhất')
    plt.title('Sự thay đổi của f(x) qua các thế hệ')
    plt.show()
    
    return best_individual, best_fitness

# Chạy thuật toán
best_x, best_f = genetic_algorithm_example1()
print(f'\nKết quả cuối cùng: x = {best_x:.4f}, f(x) = {best_f:.4f}')
````

# Thay đổi tham số thuật toán di truyền và phân tích tốc độ hội tụ

**Mục tiêu:** Thử thay đổi các tham số chính của thuật toán di truyền gồm:

- Kích thước quần thể: `pop_size`
- Số thế hệ: `generations`
- Tỷ lệ đột biến: `mutation_rate`
- Tỷ lệ lai ghép: `crossover_rate`

và phân tích ảnh hưởng của từng tham số đến tốc độ hội tụ của thuật toán.

---

## Các tham số và vai trò

| Tham số         | Ý nghĩa                                  | Ảnh hưởng chính                                                                                         |
|-----------------|-----------------------------------------|--------------------------------------------------------------------------------------------------------|
| `pop_size`      | Số cá thể trong quần thể                 | Quần thể lớn giúp khám phá không gian tìm kiếm tốt hơn nhưng tốn nhiều thời gian tính toán. Quần thể nhỏ hội tụ nhanh nhưng dễ bị kẹt cực trị cục bộ. |
| `generations`   | Số vòng lặp của thuật toán               | Số thế hệ càng nhiều thì thuật toán có nhiều cơ hội cải thiện nghiệm, tuy nhiên thời gian chạy cũng tăng. |
| `mutation_rate` | Xác suất đột biến mỗi bit                | Đột biến cao giúp tăng đa dạng gene, tránh bị kẹt trong cực trị cục bộ, nhưng nếu quá cao làm quá trình tìm kiếm nhiễu loạn. |
| `crossover_rate`| Xác suất lai ghép giữa hai cá thể       | Lai ghép giúp phối hợp gen tốt, tăng khả năng tìm nghiệm tốt. Nếu quá thấp, quần thể kém đa dạng; quá cao có thể phá hủy cấu trúc tốt của cá thể. |

---

## Phân tích ảnh hưởng

- **`pop_size`**: Kích thước quần thể lớn giúp tìm kiếm toàn cục tốt hơn nhưng chậm hơn, kích thước nhỏ làm thuật toán hội tụ nhanh nhưng dễ mắc kẹt.
- **`generations`**: Số thế hệ lớn tăng khả năng cải tiến kết quả nhưng làm tăng thời gian tính toán.
- **`mutation_rate`**: Tỷ lệ đột biến vừa phải giúp duy trì đa dạng di truyền, tránh kẹt cục bộ; quá cao hoặc quá thấp đều làm giảm hiệu quả.
- **`crossover_rate`**: Tỷ lệ lai ghép hợp lý giúp khai thác tốt thông tin di truyền, tăng hiệu quả hội tụ.

---




````python
# code vào đây 
import numpy as np
import math
import random
import matplotlib.pyplot as plt

# Hàm mục tiêu
def fitness_function(x):
    return math.sin(x) + math.cos(x)

# Khởi tạo quần thể
def initialize_population(pop_size, min_val, max_val):
    return [random.uniform(min_val, max_val) for _ in range(pop_size)]

# Lựa chọn (Roulette Wheel Selection)
def select_parents(population, fitness_values):
    total_fitness = sum(fitness_values)
    if total_fitness == 0:
        return random.choice(population)
    probabilities = [f / total_fitness for f in fitness_values]
    return random.choices(population, weights=probabilities, k=1)[0]

# Lai ghép
def crossover(parent1, parent2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        return (parent1 + parent2) / 2
    return parent1

# Đột biến
def mutate(individual, mutation_rate=0.05, min_val=0, max_val=2*math.pi):
    if random.random() < mutation_rate:
        return random.uniform(min_val, max_val)
    return individual

# Thuật toán di truyền với tham số tùy chỉnh
def genetic_algorithm(pop_size=50, generations=100, mutation_rate=0.05, crossover_rate=0.8, min_val=0, max_val=2*math.pi):
    population = initialize_population(pop_size, min_val, max_val)
    best_fitness_history = []

    for generation in range(generations):
        fitness_values = [fitness_function(x) for x in population]
        best_fitness = max(fitness_values)
        best_fitness_history.append(best_fitness)

        new_population = []
        for _ in range(pop_size):
            parent1 = select_parents(population, fitness_values)
            parent2 = select_parents(population, fitness_values)
            child = crossover(parent1, parent2, crossover_rate)
            child = mutate(child, mutation_rate, min_val, max_val)
            new_population.append(child)

        population = new_population

    best_fitness = max([fitness_function(x) for x in population])
    best_individual = population[np.argmax([fitness_function(x) for x in population])]

    return best_individual, best_fitness, best_fitness_history

# Chạy với các tổ hợp tham số khác nhau để phân tích ảnh hưởng
configs = [
    {'pop_size': 20, 'generations': 50, 'mutation_rate': 0.01, 'crossover_rate': 0.9},
    {'pop_size': 50, 'generations': 50, 'mutation_rate': 0.05, 'crossover_rate': 0.8},
    {'pop_size': 100, 'generations': 50, 'mutation_rate': 0.1, 'crossover_rate': 0.7},
]

results = []
for config in configs:
    best_x, best_f, history = genetic_algorithm(**config)
    results.append((config, best_x, best_f, history))

# Vẽ đồ thị so sánh tốc độ hội tụ
for config, best_x, best_f, history in results:
    label = f"pop={config['pop_size']}, mut={config['mutation_rate']}, cross={config['crossover_rate']}"
    plt.plot(history, label=label)

plt.xlabel('Thế hệ')
plt.ylabel('Giá trị f(x) tốt nhất')
plt.title('So sánh tốc độ hội tụ theo tham số thuật toán')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

````

# Cải tiến thuật toán – Thêm phương pháp lựa chọn khác và so sánh hiệu quả

---

## Mục tiêu

- Thêm phương pháp lựa chọn mới vào thuật toán di truyền, ví dụ như **Roulette Wheel Selection** (Lựa chọn bánh xe con quay).
- So sánh hiệu quả của phương pháp lựa chọn mới với phương pháp cũ (ví dụ: Chọn ngẫu nhiên, Tournament Selection, v.v.).
- Phân tích ảnh hưởng của phương pháp lựa chọn đến tốc độ hội tụ và chất lượng nghiệm.

---

##  Các phương pháp lựa chọn phổ biến

| Phương pháp                | Mô tả                                                                                   | Ưu điểm                              | Nhược điểm                          |
|----------------------------|-----------------------------------------------------------------------------------------|------------------------------------|-----------------------------------|
| Chọn ngẫu nhiên (Random)   | Chọn cá thể ngẫu nhiên trong quần thể                                                   | Đơn giản, nhanh                    | Không ưu tiên cá thể tốt hơn      |
| Roulette Wheel Selection    | Chọn theo xác suất tỷ lệ thuận với độ thích nghi (fitness)                              | Dễ dàng, ưu tiên cá thể tốt        | Dễ bị ảnh hưởng bởi cá thể quá mạnh|
| Tournament Selection        | Chọn ngẫu nhiên một số cá thể, chọn cá thể tốt nhất trong nhóm                          | Kiểm soát đa dạng tốt              | Cần chọn kích thước nhóm phù hợp  |

##  Ý tưởng cải tiến

1. Cài đặt thêm phương pháp **Roulette Wheel Selection** trong thuật toán di truyền.
2. Chạy thử với cùng bộ tham số trên các phương pháp lựa chọn khác nhau.
3. So sánh kết quả về:
   - Tốc độ hội tụ (số thế hệ cần để đạt nghiệm tốt).
   - Giá trị nghiệm tốt nhất đạt được.
4. Phân tích ưu – nhược điểm từng phương pháp trong thực tế.

---

##  Kết quả và phân tích

- Phương pháp **Roulette Wheel** thường giúp tăng tốc độ hội tụ nhờ ưu tiên cá thể tốt.
- Tuy nhiên có thể làm giảm đa dạng gen nếu cá thể ưu tú chiếm ưu thế quá lớn, dễ rơi vào cực trị cục bộ.
- Phương pháp **Tournament** kiểm soát tốt sự đa dạng nhưng có thể chậm hơn trong một số trường hợp.
- Phương pháp **Random** đơn giản nhưng ít hiệu quả trong tối ưu nâng cao.

---


````python
# code vào đây nhé
import numpy as np
import math
import random
import matplotlib.pyplot as plt

# Hàm mục tiêu
def fitness_function(x):
    return math.sin(x) + math.cos(x)

# Khởi tạo quần thể
def initialize_population(pop_size, min_val, max_val):
    return [random.uniform(min_val, max_val) for _ in range(pop_size)]

# 1. Chọn ngẫu nhiên
def random_selection(population, fitness_values):
    return random.choice(population)

# 2. Roulette Wheel Selection
def roulette_wheel_selection(population, fitness_values):
    total_fitness = sum(fitness_values)
    if total_fitness == 0:
        return random.choice(population)
    probabilities = [f / total_fitness for f in fitness_values]
    return random.choices(population, weights=probabilities, k=1)[0]

# 3. Tournament Selection (k = 3)
def tournament_selection(population, fitness_values, k=3):
    selected = random.sample(list(zip(population, fitness_values)), k)
    selected.sort(key=lambda x: x[1], reverse=True)
    return selected[0][0]

# Lai ghép
def crossover(parent1, parent2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        return (parent1 + parent2) / 2
    return parent1

# Đột biến
def mutate(individual, mutation_rate=0.05, min_val=0, max_val=2*math.pi):
    if random.random() < mutation_rate:
        return random.uniform(min_val, max_val)
    return individual

# Thuật toán di truyền với lựa chọn đầu vào phương pháp chọn
def genetic_algorithm(selection_method, pop_size=50, generations=100,
                      mutation_rate=0.05, crossover_rate=0.8,
                      min_val=0, max_val=2*math.pi):
    population = initialize_population(pop_size, min_val, max_val)
    best_fitness_history = []

    for generation in range(generations):
        fitness_values = [fitness_function(x) for x in population]
        best_fitness = max(fitness_values)
        best_fitness_history.append(best_fitness)

        new_population = []
        for _ in range(pop_size):
            parent1 = selection_method(population, fitness_values)
            parent2 = selection_method(population, fitness_values)
            child = crossover(parent1, parent2, crossover_rate)
            child = mutate(child, mutation_rate, min_val, max_val)
            new_population.append(child)

        population = new_population

    best_fitness = max([fitness_function(x) for x in population])
    best_individual = population[np.argmax([fitness_function(x) for x in population])]
    return best_individual, best_fitness, best_fitness_history

# Cấu hình chung
configs = {
    'pop_size': 50,
    'generations': 100,
    'mutation_rate': 0.05,
    'crossover_rate': 0.8,
}

# So sánh các phương pháp lựa chọn
methods = {
    "Random": random_selection,
    "Roulette Wheel": roulette_wheel_selection,
    "Tournament": tournament_selection
}

results = []
for name, method in methods.items():
    best_x, best_f, history = genetic_algorithm(
        selection_method=method,
        **configs
    )
    results.append((name, best_x, best_f, history))

# Vẽ đồ thị
for name, best_x, best_f, history in results:
    plt.plot(history, label=f"{name} (Best f={best_f:.4f})")

plt.xlabel('Thế hệ')
plt.ylabel('Giá trị f(x) tốt nhất')
plt.title('So sánh các phương pháp lựa chọn trong thuật toán di truyền')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
````

# Bài tập 5
# Trực quan hóa quần thể – Vẽ biểu đồ phân tán (scatter plot) thể hiện sự phân bố của quần thể qua các thế hệ

---

##  Mục tiêu

- Trực quan hóa sự phân bố của quần thể cá thể trong quá trình tiến hóa của thuật toán di truyền.
- Dùng biểu đồ phân tán (scatter plot) để thể hiện vị trí các cá thể trên không gian biến số qua từng thế hệ.
- Giúp quan sát rõ tiến trình hội tụ, đa dạng di truyền, và sự thay đổi quần thể theo thời gian.

---

## Ý tưởng thực hiện

- Với bài toán tối ưu hàm $k(x, y, z) = x^2 + y^2 + z^2$, không gian tìm kiếm là 3 chiều.
- Mỗi cá thể là một điểm trong không gian $(x, y, z)$.
- Ở mỗi thế hệ, lấy vị trí giải mã của toàn bộ cá thể, vẽ scatter plot.
- Nếu không thể trực quan 3D trực tiếp, có thể vẽ scatter 2D từng cặp biến: $(x, y)$, $(y, z)$, $(x, z)$ qua các thế hệ.

---

````python
import numpy as np
import random
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Hàm đánh giá (minimize k(x, y, z) nên dùng dấu âm)
def fitness_function(ind):
    x, y, z = ind
    return -(x**2 + y**2 + z**2)

# Khởi tạo quần thể
def initialize_population(pop_size, min_val, max_val):
    return [np.array([random.uniform(min_val, max_val) for _ in range(3)]) for _ in range(pop_size)]

# Lựa chọn roulette
def roulette_selection(population, fitness_values):
    total = sum(fitness_values)
    if total == 0:
        return random.choice(population)
    probs = [f / total for f in fitness_values]
    return random.choices(population, weights=probs, k=1)[0]

# Lai ghép trung bình
def crossover(p1, p2, rate=0.8):
    if random.random() < rate:
        return (p1 + p2) / 2
    return p1

# Đột biến 1 gen
def mutate(ind, rate=0.1, min_val=-10, max_val=10):
    if random.random() < rate:
        idx = random.randint(0, 2)
        ind[idx] = random.uniform(min_val, max_val)
    return ind

# Chạy thuật toán di truyền và lưu lịch sử
def genetic_algorithm_3d(pop_size=50, generations=50, mutation_rate=0.1,
                         crossover_rate=0.8, min_val=-10, max_val=10):
    population = initialize_population(pop_size, min_val, max_val)
    history = []

    for _ in range(generations):
        fitness_vals = [fitness_function(ind) for ind in population]
        history.append(np.array(population))  # lưu lại thế hệ
        new_pop = []

        for _ in range(pop_size):
            p1 = roulette_selection(population, fitness_vals)
            p2 = roulette_selection(population, fitness_vals)
            child = crossover(p1, p2, crossover_rate)
            child = mutate(child, mutation_rate, min_val, max_val)
            new_pop.append(child)

        population = new_pop

    return history

# Chạy thuật toán
history = genetic_algorithm_3d()

# Vẽ scatter 2D các cặp biến qua 3 thế hệ
generations_to_plot = [0, len(history)//2, len(history)-1]
titles = ['Thế hệ đầu', 'Thế hệ giữa', 'Thế hệ cuối']
pairs = [('x', 'y', 0, 1), ('y', 'z', 1, 2), ('x', 'z', 0, 2)]

fig, axs = plt.subplots(len(generations_to_plot), len(pairs), figsize=(15, 10))

for i, gen_idx in enumerate(generations_to_plot):
    gen = history[gen_idx]
    for j, (label1, label2, idx1, idx2) in enumerate(pairs):
        axs[i][j].scatter(gen[:, idx1], gen[:, idx2], alpha=0.7, c='blue')
        axs[i][j].set_xlabel(label1)
        axs[i][j].set_ylabel(label2)
        axs[i][j].set_xlim(-10, 10)
        axs[i][j].set_ylim(-10, 10)
        axs[i][j].set_title(f'{titles[i]}: {label1}-{label2}')
        axs[i][j].grid(True)

plt.tight_layout()
plt.show()

# Vẽ scatter 3D ở thế hệ cuối
final_gen = history[-1]
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(final_gen[:, 0], final_gen[:, 1], final_gen[:, 2], c='green', alpha=0.8)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title("Phân bố quần thể - Thế hệ cuối (3D)")
plt.show()
````

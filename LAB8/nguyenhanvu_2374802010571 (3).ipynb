# THỰC HÀNH GIẢI THUẬT MINIMAX - GAME XO 5x5

```python
from tkinter import *
from tkinter import messagebox

root = Tk()
root.title('Game XO NxN')

clicked = True
count = 0
win = False
buttons = []
size = 5  # Mặc định 5x5


def disableButtons():
    for row in buttons:
        for button in row:
            button.config(state=DISABLED)


def checkWinner():
    global win
    win = False

    # Hàng và cột
    for i in range(size):
        if buttons[i][0]["text"] != "" and all(buttons[i][j]["text"] == buttons[i][0]["text"] for j in range(size)):
            for j in range(size):
                buttons[i][j].config(bg="#80ffaa")
            win = True
            messagebox.showinfo("OX Game", f"Player {'1' if buttons[i][0]['text'] == 'X' else '2'} WINNER!!")
            disableButtons()
            return

        if buttons[0][i]["text"] != "" and all(buttons[j][i]["text"] == buttons[0][i]["text"] for j in range(size)):
            for j in range(size):
                buttons[j][i].config(bg="#80ffaa")
            win = True
            messagebox.showinfo("OX Game", f"Player {'1' if buttons[0][i]['text'] == 'X' else '2'} WINNER!!")
            disableButtons()
            return

    # Chéo chính
    if buttons[0][0]["text"] != "" and all(buttons[i][i]["text"] == buttons[0][0]["text"] for i in range(size)):
        for i in range(size):
            buttons[i][i].config(bg="#80ffaa")
        win = True
        messagebox.showinfo("OX Game", f"Player {'1' if buttons[0][0]['text'] == 'X' else '2'} WINNER!!")
        disableButtons()
        return

    # Chéo phụ
    if buttons[0][size - 1]["text"] != "" and all(buttons[i][size - 1 - i]["text"] == buttons[0][size - 1]["text"] for i in range(size)):
        for i in range(size):
            buttons[i][size - 1 - i].config(bg="#80ffaa")
        win = True
        messagebox.showinfo("OX Game", f"Player {'1' if buttons[0][size - 1]['text'] == 'X' else '2'} WINNER!!")
        disableButtons()
        return


def checkDraw():
    global count, win
    if count == size * size and not win:
        messagebox.showerror("OX Game", "DRAW!!")
        start()


def buttonClicked(button):
    global clicked, count
    if button["text"] == "" and clicked:
        button["text"] = "X"
        clicked = False
        count += 1
        checkWinner()
        checkDraw()
    elif button["text"] == "" and not clicked:
        button["text"] = "O"
        clicked = True
        count += 1
        checkWinner()
        checkDraw()
    else:
        messagebox.showerror("OX Game", "LỖI!! Vui lòng chọn lại")


def start():
    global buttons, clicked, count
    clicked = True
    count = 0

    # Xóa hết các button cũ
    for widget in root.winfo_children():
        if isinstance(widget, Button) and widget != startButton:
            widget.destroy()

    buttons = []
    for i in range(size):
        row = []
        for j in range(size):
            btn = Button(root, text="", font=("Helvetica", 20), height=2, width=5, bg="SystemButtonFace")
            btn.config(command=lambda b=btn: buttonClicked(b))
            btn.grid(row=i + 1, column=j)  # +1 để không đè lên dòng nhập kích thước
            row.append(btn)
        buttons.append(row)


def setSize():
    global size
    try:
        s = int(sizeEntry.get())
        if s < 3 or s > 7:
            messagebox.showerror("Lỗi", "Vui lòng nhập số từ 3 đến 10")
            return
        size = s
        start()
        sizeEntry.delete(0, END)
        sizeEntry.insert(0, str(size))
    except ValueError:
        messagebox.showerror("Lỗi", "Vui lòng nhập số hợp lệ")


# Menu
gameMenu = Menu(root)
root.config(menu=gameMenu)
optionMenu = Menu(gameMenu, tearoff=False)
gameMenu.add_cascade(label="Options", menu=optionMenu)
optionMenu.add_command(label="Restart Game", command=start)

# Dòng nhập kích thước
Label(root, text="Nhập kích thước ma trận (3-7):").grid(row=0, column=0, columnspan=2)
sizeEntry = Entry(root)
sizeEntry.grid(row=0, column=2)
sizeEntry.insert(0, str(size))
startButton = Button(root, text="Bắt đầu", command=setSize)
startButton.grid(row=0, column=3)

# Bắt đầu
start()
root.mainloop()
```
